Question 1: What is hypothesis testing in statistics?
  - Hypothesis testing is a formal statistical procedure that uses sample data to evaluate two mutually exclusive statements (hypotheses) about a population to determine which statement the data best supports. It provides a data-driven method for making inferences about a larger population and separating genuine effects from those that might occur by chance.
    - Key Components
        - Null Hypothesis(H0): This is the default assumption, typically stating there is no effect, no difference, or no relationship between variables (e.g., the average height of men is the same as the average height of women).
        - Alternative Hypothesis (Ha or H1): This is the claim the researcher is trying to prove, contradicting the null hypothesis (e.g., the average height of men is greater than that of women).
        - Significance Level(alpha): This is a pre-determined threshold (commonly 0.05 or 5%) representing the probability of rejecting the null hypothesis when it is, in fact, true (a Type I error or false positive).
        - P-value: This is the probability of observing sample results at least as extreme as those collected, assuming the null hypothesis is true. A low p-value (typically less than(alpha) provides strong evidence against the null hypothesis.
        - Test Statistic: A value calculated from the sample data(e.g., a t-score or z-score) that measures how far the sample results deviate from what is expected under the null hypothesis.
    - The Process
      - The general steps involved in hypothesis testing are:
        - State the hypotheses: Formulate both the null (H0) and alternative (Ha) hypotheses.
        - Set the significance level (alpha): Decide the acceptable risk of making a Type I error.
        - Collect and analyze data: Gather sample data and calculate the appropriate test statistic.
        - Make a decision: Compare the p-value to the significance level, or the test statistic to a critical value.
            - If the p-value is less than or equal to (alpha) (or the test statistic falls in the rejection region), you reject the null hypothesis, concluding there is enough evidence to support the alternative hypothesis.
            - If the p-value is greater than alpha (or the test statistic does not fall in the rejection region), you fail to reject the null hypothesis, concluding there is not enough evidence to support the alternative claim.
        - Interpret the results: State the conclusion in the context of the original research question.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H0) is a statement of no effect or no relationship between variables, representing the default assumption or status quo. The alternative hypothesis (Ha) or (H1) is a contradictory statement that claims a significant effect or relationship exists and is what the researcher is often trying to prove.
  - The Null Hypothesis (H0)
    - The null hypothesis is the starting point for statistical testing and is assumed to be true until enough evidence is collected to suggest otherwise, similar to assuming a defendant's innocence in a trial.
      - Core Idea: Results observed are due to chance or random causes.
      - Purpose: To provide a baseline for comparison and to be the statement that the statistical test aims to disprove.
      - Notation: Represented as \(H_{0}\) and always includes an equality symbol (e.g., (=, <=, >=).
      - Outcome: The test result is either to "reject (H0)" or "fail to reject (H0)".
    - Example: A study on a new blood pressure medicine.
      - (H0): The new medicine has no effect on blood pressure (the average blood pressure with the new drug is the same as with the old drug, Œºnew=Œºold).
  - The Alternative Hypothesis (Ha) or (H1)
    - The alternative hypothesis is the "challenger" to the null hypothesis and reflects the researcher's prediction or the effect they hope to find.
      - Core Idea: Results observed are due to a real, non-random cause.
      - Purpose: To be supported if the null hypothesis is rejected, suggesting a significant finding.
      - Notation: Represented as (Ha) or (H1) and always includes an inequality symbol (e.g.,!=, < or >).
      - Outcome: If the null hypothesis is rejected, the alternative hypothesis is considered supported.
  - Example: Continuing the medicine study.
    - (Ha): The new medicine does have an effect on blood pressure (e.g., the average blood pressure with the new drug is lower, ùúánew < ùúáold).

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - The significance level, denoted by the Greek symbol alpha (Œ±),  is a pre-determined probability threshold that defines the standard of evidence required to reject the null hypothesis in a hypothesis test. It is the maximum acceptable risk of making a Type I error (a false positive), which is the error of rejecting a true null hypothesis.
  - Significance Level in Hypothesis Testing
      - Definition: The significance level is a value chosen by the researcher before conducting the study to set the boundary for what is considered a "statistically significant" result. Common values are 0.05 (5%) and 0.01 (1%).
      - Evidentiary Standard: The Œ± level acts as an "evidentiary standard". A lower Œ± (e.g., 0.01) means the researcher requires stronger evidence against the null hypothesis to reject it, thus reducing the risk of a Type I error. A higher Œ± (e.g., 0.10) means the researcher is willing to accept weaker evidence, increasing the risk of a Type I error but also increasing the power to detect a real effect.
      - Context-Dependent: The choice of Œ± depends heavily on the consequences of a Type I error in the specific field of study. In medical research, where a false positive (claiming an ineffective drug works) can have severe consequences, a stricter Œ± (like 0.01) is often used. In exploratory research with lower stakes, a more lenient Œ± (like 0.10) might be acceptable.
  - Role in Deciding the Outcome
    - The significance level's primary role is to serve as the cut-off point for the p-value (probability value) to determine the outcome of a test. The decision rule is as follows:
      - Calculate the P-value: The p-value is the probability of observing test results at least as extreme as the actual results, assuming the null hypothesis is true.
      - Compare P-value to Œ±:
        - If P-value Œ±: The results are considered statistically significant, and the null hypothesis is rejected in favor of the alternative hypothesis. This indicates that the observed effect is unlikely to be due to random chance alone.
        - If P-value > Œ±: The results are considered not statistically significant, and the null hypothesis is not rejected (the researcher "fails to reject" it, but does not "accept" it). This means the evidence is not strong enough to conclude an effect exists.
  - In essence, the significance level defines the critical region in the sampling distribution; if the test statistic falls into this region (corresponding to a p-value below Œ±, the result is deemed sufficiently unusual to warrant rejecting the initial assumption (the null hypothesis).

Question 4: What are Type I and Type II errors? Give examples of each.
  - In statistical hypothesis testing, Type I and Type II errors are the two potential incorrect conclusions one can make when evaluating a null hypothesis (H0).
  - Type I Error
      - Definition: A Type I error, also known as a false positive or an Œ± error, occurs when a true null hypothesis is wrongly rejected. It's concluding there is a significant effect or relationship when, in reality, there isn't one.
      - Example: Medical Diagnosis
        - (H0): A person does not have a specific disease.
        - Type I Error: A medical test result says the person has the disease, but they actually do not.
        Example: Criminal Trial
          - (H0): The defendant is innocent.
          - Type I Error: The court finds the defendant guilty, but they are actually innocent.
  - ¬†Type II Error
    - ¬†Definition: A Type II error, also known as a false negative or a (Œ≤) error, occurs when a false null hypothesis is not rejected. It's failing to detect a real effect or relationship that actually exists.
    - Example:Medical Diagnosis
        -(H0): A person does not have a specific disease.
        - Type II Error: The medical test result says the person does not have the disease, but they actually do.
    - Example: Criminal Trial
        - (H0): The defendant is innocent.
        - Type II Error: The court finds the defendant not guilty(fails to reject the (H0), but they are actually guilty.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
  - The primary difference between a z-test and a t-test lies in the knowledge of the population standard deviation (œÉ) and the sample size. A z-test is used when the population standard deviation is known and the sample size is large, while a t-test is used when the population standard deviation is unknown or the sample size is small.
  - Key Differences
  1.jpg

  - When to Use Each Test
    - Use a Z-Test When:
      -The population standard deviation (œÉ) is known. This is the most crucial condition.The sample size is large (n>=30). The Central Limit Theorem ensures that with a large sample size, the sample mean will be approximately normally distributed, regardless of the population's actual distribution.
      - Example: A quality control manager at a light bulb factory knows the average lifespan of bulbs is 1,000 hours with a population standard deviation of 100 hours. They test a sample of 50 new bulbs. A z-test can determine if the new batch's average lifespan is significantly different from the known population mean.
    - Use a T-Test When:
      - he population standard deviation (œÉ) is unknown. The t-test uses the sample standard deviation (s) as an estimate.
      - The sample size is small (n<30)). The t-distribution is more spread out than the normal distribution, which accounts for the increased uncertainty that comes with a smaller sample size and an estimated population standard deviation.
      - Example: A researcher wants to test if a new fertilizer improves plant growth. They use the fertilizer on 15 plants and compare their average growth to a known mean. Since the population standard deviation for all possible plants is unknown and the sample size is small, a t-test is appropriate.
      - In modern statistical software, the t-test is often the default, as the t-distribution converges to the z-distribution as the sample size increases, making it a more flexible and robust option in practice.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# Ans
import numpy as np
from scipy import stats

def z_test_implementation(sample_data, population_mean, population_std_dev, alpha):
    """
    Performs a one-sample z-test and interprets the results.

    Args:
        sample_data (list or np.array): The sample data.
        population_mean (float): The hypothesized population mean (mu_0).
        population_std_dev (float): The known population standard deviation (sigma).
        alpha (float): The significance level (e.g., 0.05).

    Returns:
        None (prints the results and interpretation)
    """

    # Convert list to numpy array for easier calculations
    sample_data = np.array(sample_data)

    # Calculate sample statistics
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate the Z-statistic
    # Formula: Z = (sample_mean - population_mean) / (population_std_dev / sqrt(n))
    standard_error = population_std_dev / np.sqrt(n)
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value
    # For a two-tailed test, multiply by 2 (as we are interested in deviations in both directions)
    p_value = 2 * stats.norm.cdf(-abs(z_statistic))

    # Print the results
    print(f"Sample Size (n): {n}")
    print(f"Sample Mean (xÃÑ): {sample_mean:.4f}")
    print(f"Hypothesized Population Mean (Œº‚ÇÄ): {population_mean}")
    print(f"Known Population Standard Deviation (œÉ): {population_std_dev}")
    print(f"Significance Level (Œ±): {alpha}")
    print(f"\nZ-Statistic: {z_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    print("\n--- Interpretation ---")
    if p_value < alpha:
        print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis (H‚ÇÄ).")
        print(f"There is sufficient evidence to conclude that the true population mean is likely different from {population_mean}.")
    else:
        print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis (H‚ÇÄ).")
        print(f"There is not enough evidence to conclude that the true population mean is different from {population_mean}.")

# --- Example Usage ---
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume the following parameters for the hypothesis test:
# The null hypothesis is that the population mean (Œº‚ÇÄ) is 50.0
# We assume the population standard deviation (œÉ) is known to be 0.5
# We set a standard significance level (Œ±) of 0.05

hypothesized_mean = 50.0
known_pop_std_dev = 0.5
significance_level = 0.05

# Run the function
z_test_implementation(sample_data, hypothesized_mean, known_pop_std_dev, significance_level)
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# Ans
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample data.
    Uses the Student's t-distribution, which is appropriate for
    sample data where the population standard deviation is unknown.
    """
    a = 1.0 * np.array(data)
    n = len(a)
    if n <= 1:
        return np.nan, np.nan
    m, se = np.mean(a), st.sem(a)  # mean and standard error of the mean
    # Use t-distribution ppf (percent point function, inverse of cdf)
    # The value is for the alpha/2 percentile (e.g., 0.975 for 95% CI)
    h = se * st.t.ppf((1 + confidence) / 2., n - 1)
    return m, m - h, m + h

# 1. Simulate data from a normal distribution
# Define population parameters
POPULATION_MEAN = 50
POPULATION_STD = 5
SAMPLE_SIZE = 100

# Generate sample data
np.random.seed(42) # for reproducibility
data = np.random.normal(loc=POPULATION_MEAN, scale=POPULATION_STD, size=SAMPLE_SIZE)

# 2. Calculate the 95% confidence interval for the sample mean
sample_mean, ci_lower, ci_upper = calculate_confidence_interval(data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"True Population Mean: {POPULATION_MEAN}")

# 3. Plot the data using matplotlib
plt.figure(figsize=(10, 6))

# Plot a histogram of the data
plt.hist(data, bins=20, density=True, alpha=0.6, color='skyblue', label='Simulated Data Distribution')

# Plot the sample mean
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')

# Plot the confidence interval limits
plt.axvline(ci_lower, color='green', linestyle='dotted', linewidth=2, label=f'95% CI Lower Bound: {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='dotted', linewidth=2, label=f'95% CI Upper Bound: {ci_upper:.2f}')

# Highlight the confidence interval range with a shaded area
# We set the y limits to fill the entire vertical space
ymin, ymax = plt.ylim()
plt.fill_betweenx([ymin, ymax], ci_lower, ci_upper, color='green', alpha=0.1, label='95% Confidence Interval')

plt.title('Normal Distribution Data with 95% Confidence Interval for the Mean')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, axis='y', linestyle='--')
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
# Ans
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def analyze_z_scores(data, title='Distribution of Z-scores'):
    """
    Calculates z-scores from a dataset and visualizes the standardized data
    using a histogram.

    Args:
        data (array-like): The input dataset (e.g., a list or NumPy array).
        title (str): Title for the histogram plot.
    """
    # Convert data to a NumPy array for consistent handling
    data = np.array(data)

    # Calculate the Z-scores
    # ddof=0 is used for population standard deviation, which is standard for z-scores
    z_scores = zscore(data, ddof=0)

    # Print a few example z-scores and their interpretation
    print(f"Mean of original data: {np.mean(data):.2f}")
    print(f"Standard deviation of original data: {np.std(data):.2f}\n")
    print("Example Z-scores:")
    for i in range(min(5, len(z_scores))):
        print(f"  Value {data[i]:.2f}: Z-score = {z_scores[i]:.2f} (This value is {abs(z_scores[i]):.2f} standard deviations {'above' if z_scores[i] > 0 else 'below' if z_scores[i] < 0 else 'at'} the mean)")

    # Visualize the standardized data using a histogram
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel('Z-score (Number of Standard Deviations from the Mean)')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (Z-score = 0)')
    plt.legend()
    plt.grid(axis='y', alpha=0.7)
    plt.show()

    return z_scores

# Example Usage:
# Generate sample data (e.g., student test scores)
sample_data = [65, 70, 72, 75, 78, 80, 81, 83, 85, 88, 90, 92, 95, 105, 120]

# Run the function
standardized_scores = analyze_z_scores(sample_data, title='Histogram of Standardized Test Scores')
```
